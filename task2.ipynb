```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
data=pd.read_csv(r"C:\Users\Es\Downloads\Titanic-Dataset (1).csvsam")
data.head(5)
```

```python
encoder=OneHotEncoder(sparse_output=False)
one=encoder.fit_transform(data[['Embarked']].head(12))
```

```python
one
```

```python
data['Embarked']=data['Embarked'].astype(str)
```

```python
feature_names = encoder.get_feature_names_out(['Embarked'])
feature_names
```

```python
data_encoder=pd.DataFrame(one,columns=feature_names)
```

```python
data_encoder
```

```python
ind=data.columns.get_loc('Embarked')
ind
```

```python
before = data.iloc[:, :ind]
after = data.iloc[:, ind+1:]

data_final = pd.concat([before, data_encoder, after], axis=1)
data_final.head(10)
```
